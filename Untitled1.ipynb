```python
def compareConsommationBetweenTwoSeparateYears(year1,year2):
    # 1. Filtrer les années year1 et year2
    data_year1 = data[data['Année'] == year1]
    data_year2 = data[data['Année'] == year2]


    # 2. Identifier les communes communes aux deux années
    communes_communes = set(data_year1['Code Commune']).intersection(set(data_year2['Code Commune']))

    # 3. Filtrer les données pour ne garder que ces communes
    data_common = data[data['Code Commune'].isin(communes_communes) & data['Année'].isin([year1, year2])]

    # 4. Calculer les moyennes par année
    result = data_common.groupby('Année').agg({
        'Consommation annuelle totale de l\'adresse (MWh)': 'mean',
        'Consommation annuelle moyenne par logement de l\'adresse (MWh)': 'mean',
        'Consommation annuelle moyenne de la commune (MWh)': 'mean'
    }).reset_index()

    # 5. Renommer les colonnes
    result = result.rename(columns={
        'Consommation annuelle totale de l\'adresse (MWh)': 'Moyenne totale (MWh)',
        'Consommation annuelle moyenne par logement de l\'adresse (MWh)': 'Moyenne par logement (MWh)',
        'Consommation annuelle moyenne de la commune (MWh)': 'Moyenne communale (MWh)'
    })

    # 6. Afficher le résultat
    print(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def compareConsommationAcrossYears():
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    data_filtered = data[data['Année'].isin(years)]
    
    # Find communes present in all years
    commune_sets = [set(data_filtered[data_filtered['Année'] == year]['Code Commune']) 
                   for year in years]
    common_communes = set.intersection(*commune_sets)
    
    print(len(common_communes))
    data_common = data_filtered[data_filtered['Code Commune'].isin(common_communes)]
    
    # Calculate mean consumption per year
    result = data_common.groupby('Année').agg({
        'Consommation annuelle totale de l\'adresse (MWh)': 'mean',
        'Consommation annuelle moyenne par logement de l\'adresse (MWh)': 'mean',
        'Consommation annuelle moyenne de la commune (MWh)': 'mean'
    }).reset_index()
    
    # Rename columns
    result = result.rename(columns={
        'Consommation annuelle totale de l\'adresse (MWh)': 'Moyenne totale (MWh)',
        'Consommation annuelle moyenne par logement de l\'adresse (MWh)': 'Moyenne par logement (MWh)',
        'Consommation annuelle moyenne de la commune (MWh)': 'Moyenne communale (MWh)'
    })
    
    # Create histograms
    plt.figure(figsize=(15, 10))
    
    # 1. Total Consumption Histogram
    plt.subplot(3, 1, 1)
    plt.bar(result['Année'], result['Moyenne totale (MWh)'], color='skyblue')
    plt.title('Consommation Moyenne Totale par Année')
    plt.ylabel('MWh')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # 2. Per-Housing Consumption Histogram
    plt.subplot(3, 1, 2)
    plt.bar(result['Année'], result['Moyenne par logement (MWh)'], color='salmon')
    plt.title('Consommation Moyenne par Logement par Année')
    plt.ylabel('MWh')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # 3. Municipal Average Consumption Histogram
    plt.subplot(3, 1, 3)
    plt.bar(result['Année'], result['Moyenne communale (MWh)'], color='lightgreen')
    plt.title('Consommation Moyenne Communale par Année')
    plt.xlabel('Année')
    plt.ylabel('MWh')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
    
    return result
```

```python
def plot_topN_municipal_consumption(xa):
    years = [2018, 2019, 2020, 2021, 2022, 2023]
    
    # Get top 3 communes by municipal consumption
    top_communes = (data[data['Année'].isin(years)]
                   .groupby('Code Commune')['Consommation annuelle moyenne de la commune (MWh)']
                   .mean()
                   .nlargest(xa)
                   .index.tolist())
    
    # Get commune names
    commune_names = data[data['Code Commune'].isin(top_communes)][['Code Commune', 'Nom Commune']].drop_duplicates()
    name_dict = dict(zip(commune_names['Code Commune'], commune_names['Nom Commune']))
    
    # Filter and plot
    filtered_data = data[(data['Code Commune'].isin(top_communes)) & 
                       (data['Année'].isin(years))]
    
    pivot_data = filtered_data.groupby(['Année', 'Code Commune'])['Consommation annuelle moyenne de la commune (MWh)'].mean().unstack()
    
    plt.figure(figsize=(12, 6))
    for commune in pivot_data.columns:
        plt.plot(pivot_data.index, pivot_data[commune], 
                marker='o', 
                linewidth=2,
                label=name_dict.get(commune, commune))
    
    plt.title('Top '+xa+' Communes - Consommation Moyenne Communale (2018-2023)')
    plt.xlabel('Année')
    plt.ylabel('MWh')
    plt.xticks(years)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(title='Commune')
    
    # Add trend lines
    for commune in pivot_data.columns:
        z = np.polyfit(pivot_data.index, pivot_data[commune], 1)
        p = np.poly1d(z)
        plt.plot(pivot_data.index, p(pivot_data.index), "--", alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```
