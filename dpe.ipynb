```python
# Sélection des colonnes à analyser
columns_to_check = [
    'Conso_5_usages_é_finale_énergie_n°1',
    'Conso_chauffage_é_finale_énergie_n°1',
    'Conso_chauffage_é_primaire',
    'Conso_5_usages_é_primaire'
]
```

```python
# 📊 Visualisation initiale : Histogrammes et Boxplots**
def plot_distribution(data, columns):
    plt.figure(figsize=(15, 8))
    for i, col in enumerate(columns):
        plt.subplot(2, len(columns), i+1)
        sns.histplot(data[col], bins=30, kde=True, color='blue')
        plt.title(f"Histogramme de {col}")

        plt.subplot(2, len(columns), i+1+len(columns))
        sns.boxplot(x=data[col], color='red')
        plt.title(f"Boxplot de {col}")

    plt.tight_layout()
    plt.show()

plot_distribution(dpe_existants2, columns_to_check)
```

📊 Partie supérieure : Histogrammes
Les histogrammes montrent des distributions très asymétriques à droite (right-skewed), avec une majorité de valeurs faibles, mais quelques cas avec des valeurs très élevées (longue queue).

Cela suggère la présence de valeurs extrêmes ou outliers qui tirent la distribution vers la droite.

Le pic très élevé à gauche indique que la majorité des logements consomment peu d'énergie, ce qui est attendu, mais certains cas ont des consommations anormalement élevées

📦 Partie inférieure : Boxplots
Les 4 graphiques du bas montrent les boxplots correspondants à chaque variable ci-dessus.

Les boxplots confirment la présence de nombreux outliers (points isolés au-delà des moustaches).

La boîte (entre le 1er et le 3e quartile) est très étroite, ce qui signifie que la majorité des valeurs sont regroupées vers le bas.

Les moustaches sont très courtes par rapport à la gamme totale des données, ce qui renforce l’idée que quelques points anormaux influencent fortement la distribution.

```python
#  Détection des outliers avec IQR**
def detect_outliers_iqr(data, column, threshold=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Test avec IQR = 1.5 et 2.5
for col in columns_to_check:
    outliers_1_5, lb1, ub1 = detect_outliers_iqr(dpe_existants2, col, threshold=1.5)
    outliers_2_5, lb2, ub2 = detect_outliers_iqr(dpe_existants2, col, threshold=2.5)
    
    print(f"\n📌 **{col}**")
    print(f"🔹 Outliers avec IQR=1.5 : {len(outliers_1_5)}")
    print(f"🔹 Outliers avec IQR=2.5 : {len(outliers_2_5)}")
```

```python
#  **Méthode alternative : Z-score**
import numpy as np
from scipy.stats import zscore
def detect_outliers_zscore(data, column, threshold=3):
    z_scores = np.abs(zscore(data[column]))
    return data[z_scores > threshold]

for col in columns_to_check:
    outliers_z = detect_outliers_zscore(dpe_existants2, col)
    print(f"🔹 Outliers avec Z-score > 3 pour {col} : {len(outliers_z)}")
```

```python
# **Traitement des outliers**
import numpy as np
def remove_outliers(data, column, method='IQR', threshold=1.5):
    if method == 'IQR':
        _, lb, ub = detect_outliers_iqr(data, column, threshold)
        return data[(data[column] >= lb) & (data[column] <= ub)]
    elif method == 'Z-score':
        z_scores = np.abs(zscore(data[column]))
        return data[z_scores <= threshold]
    return data
```

```python
# Suppression des outliers avec IQR = 2.5
dpe_cleaned = dpe_existants2.copy()
for col in columns_to_check:
    dpe_cleaned = remove_outliers(dpe_cleaned, col, method='IQR', threshold=2.5)

print(f"\n✅ Nombre de lignes après suppression des outliers : {len(dpe_cleaned)}")

# Visualisation après traitement
plot_distribution(dpe_cleaned, columns_to_check)
```

### Corrélations entre les variables (cohérence des liens)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def create_heatmap(df, figsize=(12, 10)):
    # Convert object columns to numeric
    df_numeric = df.copy()
    object_cols = df_numeric.select_dtypes(include=['object']).columns
    # Convert object columns to categorical codes
    for col in object_cols:
        df_numeric[col] = pd.Categorical(df_numeric[col]).codes
    
    # Compute and visualize correlation matrix
    plt.figure(figsize=figsize)
    sns.heatmap(
        df_numeric.corr(), 
        annot=True, 
        cmap='coolwarm', 
        center=0, 
        fmt='.2f',
        linewidths=0.5,
        cbar_kws={"shrink": .8}
    )
    
    plt.title('Correlation Heatmap', fontsize=15)
    plt.tight_layout()
    return plt

# Usage example
heatmap = create_heatmap(dpe_cleaned)
heatmap.show()
```

```python
# Liste des types de chauffage électriques
types_electriques = ['Électricité', "Électricité d'origine renouvelable utilisée dans le bâtiment"]

# Ajouter une colonne Chauffage_électrique (1 si électrique, 0 sinon)
dpe_cleaned['Chauffage_électrique'] = dpe_cleaned['Type_énergie_principale_chauffage'].isin(types_electriques).map({True: "Oui", False: "Non"})
dpe_cleaned
```
