```python
# Rename columns to match between datasets
dpe_existants = dpe_existants.rename(columns={'Nom__commune_(BAN)': 'Nom Commune', 'N°_département_(BAN)': 'Code Département'})
Consommation_électrique = Consommation_électrique.rename(columns={'Commune': 'Nom Commune', 'Département': 'Code Département'})
temperature = temperature.rename(columns={'Département': 'Code Département', 'Commune': 'Nom Commune'})
```

```python
# Convert 'Code Département' to string for all DataFrames
dpe_existants["Code Département"] = dpe_existants["Code Département"].astype(str)
Consommation_électrique["Code Département"] = Consommation_électrique["Code Département"].astype(str)
temperature["Code Département"] = temperature["Code Département"].astype(str)

# Convert 'Nom Commune' to string for all DataFrames
dpe_existants["Nom Commune"] = dpe_existants["Nom Commune"].astype(str)
Consommation_électrique["Nom Commune"] = Consommation_électrique["Nom Commune"].astype(str)
temperature["Nom Commune"] = temperature["Nom Commune"].astype(str)
```

```python
# Merge DPE with Consommation Électrique
df_merged = dpe_existants.merge(Consommation_électrique, on=['Nom Commune', 'Code Département'], how='inner')

df_merged
```

```python
# Merge with Température DataFrame
df_final = df_merged.merge(temperature, on=['Nom Commune', 'Code Département'], how='inner')

# Display result
df_final
```

```python
dpe_existants["Nom Commune"] = dpe_existants["Nom Commune"].str.lower().str.strip()
Consommation_électrique["Nom Commune"] = Consommation_électrique["Nom Commune"].str.lower().str.strip()
temperature["Nom Commune"] = temperature["Nom Commune"].str.lower().str.strip()
```

```python
# Convert all department codes to string and remove decimals
dpe_existants["Code Département"] = dpe_existants["Code Département"].astype(str).str.zfill(2)

Consommation_électrique["Code Département"] = (
    Consommation_électrique["Code Département"]
    .astype(str)  # Convert everything to string
    .str.replace(r"\.0$", "", regex=True)  # Remove .0 if present
    .str.zfill(2)  # Ensure two-digit format
)
temperature["Code Département"] = (
    temperature["Code Département"]
    .astype(str)  # Convert everything to string
    .str.replace(r"\.0$", "", regex=True)  # Remove .0 if present
    .str.zfill(2)  # Ensure two-digit format
)
```

```python
import sys
sys.path.append(r"c:\users\msi\appdata\roaming\python\python310\site-packages")
import unidecode

# Function to clean commune names
def clean_commune(name):
    if isinstance(name, str):
        name = name.lower().strip()  # Lowercase & strip spaces
        name = unidecode.unidecode(name)  # Remove accents
        name = name.replace("-", " ")  # Replace hyphens with spaces
    return name

# Apply cleaning function to all datasets
dpe_existants["Nom Commune"] = dpe_existants["Nom Commune"].apply(clean_commune)
Consommation_électrique["Nom Commune"] = Consommation_électrique["Nom Commune"].apply(clean_commune)
temperature["Nom Commune"] = temperature["Nom Commune"].apply(clean_commune)
```

```python
def optimize_memory(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = df[col].astype('float32')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = df[col].astype('int32')
    for col in df.select_dtypes(include=['object']).columns:
        if df[col].nunique() / len(df[col]) < 0.5:  # Si peu de valeurs uniques
            df[col] = df[col].astype('category')
    return df

dpe_existants = optimize_memory(dpe_existants)
Consommation_électrique = optimize_memory(Consommation_électrique)
```

```python
# Merge DPE with Consommation Électrique
df_merged = dpe_existants.merge(Consommation_électrique, on=['Nom Commune', 'Code Département'], how='inner')
```
