```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dpe_existants = pd.read_csv('dpe_cleaned-final.csv')
temperature = pd.read_csv('temperature_27_03.csv')
Consommation_électrique = pd.read_csv('consommation-annuelle-residentielle-par-adresse.csv',sep=';')
```

```python
# Rename columns to match between datasets
dpe_existants = dpe_existants.rename(columns={'Nom__commune_(BAN)': 'Nom Commune', 'N°_département_(BAN)': 'Code Département'})
Consommation_électrique = Consommation_électrique.rename(columns={'Commune': 'Nom Commune', 'Département': 'Code Département'})
temperature = temperature.rename(columns={'Département': 'Code Département', 'Commune': 'Nom Commune'})
dpe_existants = dpe_existants.rename(columns={'Adresse_brute': 'Adresse'})
```

```python
print("\nUnique 'Nom Commune' in DPE:", dpe_existants['Nom Commune'].unique()[:10])
print("Unique 'Nom Commune' in Conso:", Consommation_électrique['Nom Commune'].unique()[:10])
print("Unique 'Nom Commune' in Température:", temperature['Nom Commune'].unique()[:10])

print("\nUnique 'Code Département' in DPE:", dpe_existants['Code Département'].unique()[:10])
print("Unique 'Code Département' in Conso:", Consommation_électrique['Code Département'].unique()[:10])
print("Unique 'Code Département' in Température:", temperature['Code Département'].unique()[:10])
```

```python
print("Missing values in DPE:")
print(dpe_existants[['Nom Commune', 'Code Département']].isnull().sum())

print("\nMissing values in Consommation Électrique:")
print(Consommation_électrique[['Nom Commune', 'Code Département']].isnull().sum())

print("\nMissing values in Température:")
print(temperature[['Nom Commune', 'Code Département']].isnull().sum())
```

```python
dpe_existants["Nom Commune"] = dpe_existants["Nom Commune"].str.lower().str.strip()
Consommation_électrique["Nom Commune"] = Consommation_électrique["Nom Commune"].str.lower().str.strip()
temperature["Nom Commune"] = temperature["Nom Commune"].str.lower().str.strip()
```

```python
# Convert 'Code Département' to string for all DataFrames
dpe_existants["Code Département"] = dpe_existants["Code Département"].astype(str)
Consommation_électrique["Code Département"] = Consommation_électrique["Code Département"].astype(str)
temperature["Code Département"] = temperature["Code Département"].astype(str)

# Convert 'Nom Commune' to string for all DataFrames
dpe_existants["Nom Commune"] = dpe_existants["Nom Commune"].astype(str)
Consommation_électrique["Nom Commune"] = Consommation_électrique["Nom Commune"].astype(str)
temperature["Nom Commune"] = temperature["Nom Commune"].astype(str)
```

```python
print(Consommation_électrique['Année'].value_counts())
```

```python
# Merge DPE with Consommation Électrique
df_merged = dpe_existants.merge(Consommation_électrique, on=['Adresse',"Nom Commune"], how='inner')

selected = df_merged.sort_values(["Nom Commune","Adresse"])

selected2 = df_merged.value_counts(["Année"])
selected2
```
